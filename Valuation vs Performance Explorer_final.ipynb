```python
import pandas as pd

# Step 1: Load the CSV
df = pd.read_csv(r'C:\Users\alire\sp500_project\sp500_companies.csv')

# Step 2: Convert columns to numeric where needed
cols_to_convert = ['Marketcap', 'Ebitda', 'Revenuegrowth']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Step 3: Drop rows missing important data
df_clean = df.dropna(subset=['Sector', 'Marketcap', 'Ebitda', 'Revenuegrowth'])

print(df_clean.head())
```

```python
# Step 3: Sector-Level Grouping & Aggregation

sector_summary = df_clean.groupby('Sector').agg({
    'Marketcap': 'sum',             # Total market cap per sector
    'Ebitda': 'mean',               # Average EBITDA
    'Revenuegrowth': 'mean',        # Average revenue growth
    'Symbol': 'count'               # Number of companies
}).rename(columns={
    'Marketcap': 'Total_Marketcap',
    'Ebitda': 'Avg_Ebitda',
    'Revenuegrowth': 'Avg_RevenueGrowth',
    'Symbol': 'Company_Count'
}).reset_index()

# Sort by total market cap
sector_summary = sector_summary.sort_values(by='Total_Marketcap', ascending=False)

# Display the summary
print(sector_summary)
```

```python
# Total market cap (to calculate percentage share)
total_marketcap = sector_summary['Total_Marketcap'].sum()

# Add % Market Share and Sector Ranking
sector_summary['Marketcap_Share_%'] = (sector_summary['Total_Marketcap'] / total_marketcap) * 100
sector_summary['Sector_Rank'] = sector_summary['Total_Marketcap'].rank(ascending=False).astype(int)

# Optional: Format as percent (only for display)
sector_summary['Avg_RevenueGrowth_%'] = sector_summary['Avg_RevenueGrowth'] * 100
```

```python
# Save final version for Tableau
sector_summary.to_csv(r'C:\Users\alire\sp500_project\sector_summary_for_tableau.csv', index=False)
```

```python
import pandas as pd
import yfinance as yf

# Step 1: Get all S&P 500 tickers from Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table = pd.read_html(url)
sp500_table = table[0]
tickers = sp500_table['Symbol'].tolist()

# Step 2: Fetch data for each ticker
data = []
for ticker in tickers:
    try:
        stock = yf.Ticker(ticker)
        info = stock.info
        data.append({
            'Symbol': ticker,
            'Sector': info.get('sector'),
            'Marketcap': info.get('marketCap'),
            'Revenuegrowth': info.get('revenueGrowth'),
            'PE': info.get('trailingPE'),
            'PEG': info.get('pegRatio'),
            'Longname': info.get('longName')
        })
    except Exception as e:
        print(f"Skipping {ticker} due to error: {e}")

# Step 3: Convert to DataFrame and save
df = pd.DataFrame(data)
df.to_csv("full_sp500_valuation_performance.csv", index=False)

print("Data saved to full_sp500_valuation_performance.csv")
```
